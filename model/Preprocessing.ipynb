```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv('data/tmdb_5000_credits.csv')
movies  = pd.read_csv('data/tmdb_5000_movies.csv')
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head(2)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(2)
```

```python
movies.isna().sum()
```

```python
movies= movies.dropna()
```

```python
movies.isna().sum()
```

```python
movies.shape
```

```python
import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(5)
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
movies['cast'] = movies['cast'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
movies['crew'] = movies['crew'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
movies['keywords'] = movies['keywords'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
```

```python
movies.head(5)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview'] = movies['overview'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] +" " +  movies['genres'] +" " +  movies['keywords']  +" " +  movies['cast']  +" " +  movies['crew']
```

```python
df = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
df.head()
```

```python
df.to_csv('data/Movies.csv')
```

```python
df['tags'] = df['tags'].str.lower()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(df['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = df[df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(df.iloc[i[0]].title)
```
